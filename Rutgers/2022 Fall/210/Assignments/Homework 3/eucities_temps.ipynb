# EU Cities Temperatures
## Preprocessing / analysis
1. Fill in the missing latitude and longitude values by calculating the average for that country. Round the average to 2 decimal places.

```python
import pandas as pd
import numpy as np

f = pd.read_csv("EuCitiesTemperatures.csv")
countries = f.groupby("country")
means = countries[["latitude","longitude"]].mean()
# print(means["latitude"])

f["latitude"].fillna(value=means["latitude"], inplace=True)
def fx(row):
    if np.isnan(row["latitude"]):
        row["latitude"] = means[row["country"]]
    return row


f = f.apply(fx, 1)
# print(f)

f.to_csv("updatedTemps.csv")
```
